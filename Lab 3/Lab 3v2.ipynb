```python
# process of cleaning xmas before constructing the complete function. This code chunk will be included as part of the final function. 
# assigns correct string type variables to verb, adjective and location, fills the na values in these variables will blanks

#xmas['Verb'] = xmas['Verb'].astype(str)
#xmas['Adjective'] = xmas['Adjective'].astype(str)
#xmas['Location'] = xmas['Location'].astype(str)
#xmas['Verb'] = xmas['Verb'].fillna('')
#xmas['Adjective'] = xmas['Adjective'].fillna('')
#xmas['Location'] = xmas['Location'].fillna('')

def clean_dataset(dataset):
    """
    clean_columns function takes a dataset with the columns named in the same format as xmas, and cleans it
    """
    clean_columns = ['Verb', 'Adjective', 'Location']
    
    for column in clean_columns: 
        dataset[column] = dataset[column].astype(str)
        dataset[column] = dataset[column].fillna('')
    return dataset
```

#### Function 1: pluralize_gift() <br>
The gifts are listed in singular: for example, on day five the narrator receives “five golden rings”, but the entry in the data set for the gift on day five simply says “ring”. <br>
Using the skeleton of the pluralize_gift() function, complete the code so that the function takes a gift and returns the appropriate plural.

```python
#Define a function with 3 conditions according to rules: ies, ee, and s as only plural forms
#take the singular gift for from dataset, OR ANY DATASET, and return the plural values 

def pluralize_gift(gift):
    """
    gift (string): argument for pluralize_gift. It is the string we intend to pluralize with the given plural rules
    gift.endswith pandas method referenced from Stack Overflow 
    gift[:-1]: slices string and removes the last character (y)
    """
    if gift.endswith('y'):
        gift = gift[:-1] + 'ies'
    elif 'oo' in gift:
        gift = gift.replace('oo', 'ee')
    else: 
        gift += 's'
    return gift
    
```

#### Function 2: make_phrase() <br>
Write a function called make_phrase() that takes as input the necessary information, and returns a phrase.

```python
def make_phrase(day, day_word, gift, verb, adjective, location):
    """
    day (integer): represents the day associated with the other arguments. Acts similarly to an index of the dataset. 
    day_word (string): the day as a word of ordinal value. This gets mapped to word_day, which changes the ordinal word to its literal word using map_word.
    gift (string): same as above. It is the string we intend to pluralize with the given plural rules. Only gets pluarlized if the day is greater than one.
    verb (string): The verb associated in the same index as as the gift.
    adjective (string): Adjective associated with the gift. 
    location: The location associated with the gift. 
    make_phrase (function): returns the arguments in phrase order after running the function operations. This will return the conditional values in each day, without the filler words of the complete ssong. 
    """
    
    # Step 1: Replace NAs with blank strings
    verb = verb if verb != "nan" else ''
    adjective = adjective if adjective != "nan" else ''
    location = location if location != "nan" else ''
    
    map_day =  {
    'first': 'one',
    'second': 'two',
    'third': 'three',
    'fourth': 'four',
    'fifth': 'five',
    'sixth': 'six',
    'seventh': 'seven',
    'eighth': 'eight',
    'ninth': 'nine',
    'tenth': 'ten',
    'eleventh': 'eleven',
    'twelfth': 'twelve'}

    word_day = map_day.get(day_word)
    ## Step 2: If the day number is larger than 1, the gift items need pluralized!
    if day > 1:
        gift = pluralize_gift(gift)

    ## Step 3: Figure out if a gift item starts with a vowel
    if isinstance(gift, str) and len(gift) > 0:
        vowels = ['a', 'e', 'i', 'o', 'u']
        bvowel = gift[0].lower() in vowels #.lower insures that the letters in gift are correctly mapped to the vowels list. (converts the letters to lowercase before mapping)
    else: 
        bvowel = False
    ## Step 4: For the first day, handle the 'a' or 'an' condition
    if day == 1:
        if bvowel:
            day = 'an'
        else:
            day = 'a'
    else:
        day = word_day

    ## Step 5: Construct the final phrase and return it
    phrase = f"{day} {adjective} {gift} {verb} {location}".strip()
    return phrase
```

```python
# Apply the function and print the results
# Function application creates a new column in the dataset called "Full Phrase". This column applies the phrase calculated in the function make phrase, for each day and its associated variables. 
# Simiilar to above, this code chunk will be included in the final function. 
# lambda row: referenced from ChatGPT: defines an anonymous function that takes a single row argument, representing each row in the dataset. This simplifies the process of applying the make phrase function to any dataset with the same structure as xmas.
xmas['Full Phrase'] = xmas.apply(
    lambda row: make_phrase(
        day = row["Day"],
        day_word = row["Day.in.Words"],
        gift = row["Gift.Item"],
        verb = row["Verb"],
        adjective = row["Adjective"],
        location = row["Location"]
    ),
    axis = 1
)

xmas['Full Phrase'].fillna('', inplace = True)
print(xmas[[ "Full Phrase"]])
```

#### Function 3: sing_day() <br>
Write a function called sing_day() that takes as input:

A dataset (input as a dataframe)

A number indicating which day to sing about (input as an integer)

The name of a column in the dataset that contains the phrases for each day (input as an tidy name)

```python
def sing_day(dataset, num, phrase_col):
  """
  function sing_day uses filler words to sing the phrase asssociated with the day. 
  dataset (pandas dataframe): contains information about each day's gift, most importantly the phrase_col
  num (integer): represents the number associated with phrase_col, or the the row in the dataset. 
  phrase_col (string): column in dataset that contains the complete phrase
  map_num_to_word: key map used to get the rspective num in literal word form. This mapped value takes on the variable num_word. 
  intro (string): This is the full intro string with respective num_word included for each verse 
  gifts (empty string): iterative loop that assigns the correct order (descending) in which the gifts are sung in sing_day
  range(num, 0, -1): num is starting point. 0 is stopping point. -1 is the step taken. This means that the loop goes from num to 0, decreasing by -1.
  if i == 1 and num !=1: checks if current iteration is 1, and num is not also 1. This is needed so the resulting first line in the song is distinct from the rest of the song. 
  \n: line break for each phrase in the iteration 
  gifts.strip(): referenced from Gemini: cleans up the string by eliminating unwanted spaces or newline characters. Output should not have extra spaces or blank lines.
  """
  dataset = clean_dataset(dataset)

  # Step 1: Setup the intro line
  map_num_to_word = {
    1:'first',
    2:'second',
    3:'third',
    4:'fourth',
    5:'fifth',
    6:'sixth',
    7:'seventh',
    8:'eighth',
    9:'ninth',
    10:'tenth',
    11:'eleventh',
    12:'twelvth'}

  num_word = map_num_to_word.get(num) 
  intro = "On the " + num_word + " day of Christmas, my true love sent to me:"
  
  # Step 2: Sing the gift phrases
  # Hint: What order are they gifts sung in each day?
  gifts = ""
  for i in range(num, 0, -1):
    if i == 1 and num != 1:
      gifts += "and " + dataset.loc[dataset['Day'] == i, phrase_col].values[0] + "\n"
    else: 
      gifts += dataset.loc[dataset['Day'] == i, phrase_col].values[0] + "\n"
      
  # Step 3: Put it all together and return
  return intro + "\n" + gifts.strip()
```

Use Your Functions!
Run appropriate code to output the lyrics for the entire 12 Days of Christmas song.

```python
# This final function sing_full_song calls the functions created above and prints the the full song using an iterative loop

def sing_full_song(dataset, phrase_col):
    """
    dataset (pandas dataframe): same as above
    phrase_col (string): same as above
    full_song (empty list): takes values from iterative list below, calling sing_day function.  
    range(1, 13): for 12 days of christmas 
    full_song.append(line): appends line on the full song.
    \n breaks line for every new line in full_song.
    clean_dataset called from earlier function
    """   
    dataset = clean_dataset(dataset)

    if phrase_col not in dataset.columns:
        dataset[phrase_col] = dataset.apply(
            lambda row: make_phrase(
                day=row["Day"],
                day_word=row["Day.in.Words"],
                gift=row["Gift.Item"],
                verb=row["Verb"],
                adjective=row["Adjective"],
                location=row["Location"]
            ),
            axis=1
        )

    full_song = []
    
    for k in range(1, 13):
        line = sing_day(dataset, k, phrase_col)
        full_song.append(line)
    print("\n\n".join(full_song))
```
